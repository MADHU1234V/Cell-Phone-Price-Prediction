```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Cellphoneprice.csv')
df.head()
```

```python
flask_features = ['ram', 'battery_power', 'px_height', 'px_width', 'three_g', 'four_g']
```

```python
# Prepare inputs and target
df_flask = df[flask_features + ['price_range']]
X = df_flask.drop('price_range', axis=1)
Y = df_flask['price_range']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state = 42)
```

```python
lr= LogisticRegression(multi_class='multinomial', solver='newton-cg', max_iter=20000)
lr.fit(X_train, Y_train)
```

```python
Y_pred_lr = lr.predict(X_test)
```

```python
Y_pred_lr.shape
```

```python
print("Accuracy:", accuracy_score(Y_test, Y_pred_lr)*100,'%')
print(classification_report(Y_test, Y_pred_lr))
```

```python
# Test prediction
sample = np.array([[2000, 3000, 1080, 1920, 1, 1]])
prediction = lr.predict(sample)[0]
print('Sample prediction:', prediction)
```

```python
import joblib
joblib.dump(lr,'CellPricePrediction.pkl')
```
