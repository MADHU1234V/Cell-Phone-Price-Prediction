## Exploratory data analysis [EDA]

```python
ax = sns.barplot(x="price_range", y="ram", data=df, ci=None, palette="viridis")
plt.title("RAM vs Price Range")
for container in ax.containers:
    ax.bar_label(container,fmt='%.0f')
    
plt.show()
```

- As the price range increases, the average RAM capacity also increases significantly.

- The chart indicates a positive correlation between price and RAM: higher-priced devices tend to have more RAM.

```python
bx=sns.barplot(y= 'battery_power', x= 'price_range', data = df,ci=None, palette="viridis")
plt.title("Battery power vs Price range")
for container in bx.containers:
    bx.bar_label(container,fmt= '%0.f')
plt.show()
```

- Battery power increases slightly with price range, but the growth is not as strong as we saw with RAM.

- From Price Range 1 to 2, the values are almost the same (~1229 vs 1228).

- The highest price range (3) has the largest battery power, ~1380 mAh.

```python
sns.scatterplot(y="ram", x="battery_power", data=df, palette = 'muted',hue= 'price_range')
plt.title("RAM vs Battery Power")
plt.show()

```

 RAM is the strongest separator of price range:

- Price Range 0 (blue) is concentrated in the lower RAM values (<1500 MB).

- Price Range 1 (orange) is between aprox 1000–2500 MB.

- Price Range 2 (green) lies between aprox 2000–3000 MB.

- Price Range 3 (red) dominates the highest RAM values (>3000 MB).

Battery power is spread more evenly:

- Across all ranges, battery power spans from aprox 500 to 2000.

- Unlike RAM, it does not clearly separate price ranges.

```python
sns.barplot(y="px_width", x="price_range", data=df, alpha=0.5)
plt.title("Pixel Width vs Price Range")
plt.show()
```

- Phones in the lowest price range have the lowest average pixel width.

- There is a steady upward trend in screen resolution as price range rises, with the highest price category (3) featuring the highest average px_width.

- This demonstrates that higher-priced phones typically offer better display quality (wider screens or higher resolution).

```python
sns.scatterplot(x="px_width", y="px_height", hue="price_range", data=df, alpha=0.5, palette="viridis")
plt.title("Screen Resolution: Width vs Height")
plt.show()
```

- Points are color-coded by price range (0 to 3), representing the progression from cheapest to costliest phones.

- Higher-priced phones (yellow, price range 3) cluster in the upper right, indicating they generally have both higher px_width and px_height, i.e., better overall screen resolution.

- Lower-priced phones (purple, price range 0) appear mostly in the lower left, showing smaller and lower-resolution screens.

- There is a clear upward trend: as the px_width increases, px_height also increases, and these higher values are dominated by higher price classes.

- Internal memory (int_memory), primary camera (pc), and pixel height (px_height) contribute modestly but not nearly as strongly as RAM and px_width.

- Many feature pairs are nearly uncorrelated (values close to zero), showing these are mostly independent characteristics.

### Checking Outliers

```python
numeric_cols = df.select_dtypes(include=['int64','float64']).columns
skewness = df[numeric_cols].skew().sort_values(ascending=False)
print(skewness)
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
def detect_outliers(df):
    for col in df.select_dtypes(include=['int64', 'float64']).columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        
        outliers = df[(df[col] < lower) | (df[col] > upper)]
        print(col, " -> Outliers:", len(outliers), 
              " (", round(len(outliers)/len(df)*100, 2), "% )")

detect_outliers(df)
```

1. three_g (23.85% outliers)

- This is a binary feature (0 or 1).

- The IQR method incorrectly flagged ~477 rows as “outliers” because all values are at the boundary (1).

- Ignore this. Outlier detection doesn’t apply to binary (0/1) features.

2. fc (0.9% outliers)

- A few phones with unusually high front camera megapixels (above 16 MP) were flagged.

- Makes sense: most phones in the dataset have small front cameras.

3. px_height (0.1% outliers)

- Only 2 rows flagged.

- Phones with very high resolution heights compared to the rest.

- Not a major issue.

4. All other features → 0% outliers

- That means their distributions are within the IQR thresholds.

- No extreme values were detected.

```python
# Winsorize selected skewed columns
for col in ["fc", "px_height"]:
    lower = df[col].quantile(0.01)
    upper = df[col].quantile(0.99)
    df[col] = df[col].clip(lower=lower, upper=upper)

print("Winsorization applied to fc and px_height.")
```

```python
def detect_outliers(df):
    for col in df.select_dtypes(include=['int64', 'float64']).columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        
        outliers = df[(df[col] < lower) | (df[col] > upper)]
        print(col, " -> Outliers:", len(outliers), 
              " (", round(len(outliers)/len(df)*100, 2), "% )")

# Example usage
detect_outliers(df)
```

# Model fitting

## Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
lr= LogisticRegression(multi_class='multinomial', solver='newton-cg', max_iter=20000)
```

```python
lr.fit(X_train, Y_train)
```

```python
Y_pred_lr = lr.predict(X_test)
```

```python
Y_pred_lr.shape
```

```python
print("Accuracy:", accuracy_score(Y_test, Y_pred_lr)*100,'%')
print(classification_report(Y_test, Y_pred_lr))
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion='log_loss',max_depth=10,min_samples_split=5,min_samples_leaf=2,random_state=133)
```

```python
dt.fit(X_train, Y_train)
```

```python
Y_pred_dt = dt.predict(X_test)
```

```python
print(f"Accuracy: {accuracy_score(Y_test, Y_pred_dt)*100}",'%')
print(classification_report(Y_test, Y_pred_dt))
```

## SVM

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
max_accuracy = 0
for x in range(2000):
    rf = RandomForestClassifier(random_state=x)
    rf.fit(X_train,Y_train)
    Y_pred_rf = rf.predict(X_test)
    current_accuracy = round(accuracy_score(Y_pred_rf,Y_test)*100,2)
    if(current_accuracy>max_accuracy):
        max_accuracy = current_accuracy
        best_x = x
```

```python
rf = RandomForestClassifier(random_state=best_x)
```

```python
rf.fit(X_train,Y_train)
```

```python
Y_pred_rf = rf.predict(X_test)
```

```python
print('Accuracy:', accuracy_score(Y_test,Y_pred_rf)*100, '%')
print(classification_report(Y_test,Y_pred_rf))
```

- The top performers in accuracy are Logistic Regression (97.6%) and SVM (97.0%).

- KNN is a close competitor with 95% accuracy.

- Tree-based methods like Random Forest and Decision Trees provide lower accuracy in this case (90.4% and 85.6% respectively)
